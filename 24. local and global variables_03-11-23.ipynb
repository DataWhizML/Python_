**Local variables**

```python
def addition():
    n1=10
    n2=20
    add=n1+n2
    print("addition of {} and {} is {}".format(n1,n2,add))
```

**local variables**: 
- variables provided inside the function
**global variables**:
- variables provided outside the function

```python
 n1=10
 n2=20
def addition():
    add=n1+n2
    print("addition of {} and {} is {}".format(n1,n2,add))
addition()
```

```python
def avg():
    n1=20
    n2=30
    n3=40
    avg=(n1+n2+n3)/3
    print(avg)
avg()
#Local
```

```python
- local variables are inside the function
- local variable values cannot be used outside the function untill and unless you return those values.
```

**Global Variables**

```python
n1=20
n2=30
n3=40
def avg():
    avg=(n1+n2+n3)/3
    print(avg)
avg()
#Global
```

```python
n1=20
n2=30
def addition(n1):
    n1=200
    n2=300
    add=n1+n2
    print(add)
addition(500)
print(n1)
```

```python
n11=20
n22=30
def addition123(n11=700):
    n22=300
    add12=n11+n22
    print("addition is",add12)
addition123(500)
print(n11)
```

```python
#we cannot use local variable values outside the function until and unless return statement is there
#use global keyword to use local variable outside a function
```

```python
#without return we need to get the updated value outside the function
#using the local variable outside the function
num1=10
def updated():
    global num1,num2
    num2=20
    num1= num1+100
    print("num1 value inside function is ",num1)
updated()#110
print(num1)#10
print(num2)
```

```python
n_1=10
def updated_():
    global n_1
    n_1=20
    n_1=n_1+100
    print("n1 value inside function",n_1)
updated_()
print(n_1)
```

```python
 n111=20
 n222=30
def addition(n111):
    global n222,add123
    n222=300
    add123=n111+n222
    print("addition of {} and {} is {}".format(n111,n222,add123))
addition(500)
print(n222)
print(add123)
```
